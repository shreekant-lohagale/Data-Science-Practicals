## Encoding Categorical Feature

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Sample dataset
df = pd.DataFrame({
    'Name': ['Ravi', 'Sneha', 'Amit', 'Pooja'],
    'Gender': ['Male', 'Female', 'Male', 'Female']
})

# Apply Label Encoding
le = LabelEncoder()
df['Gender_Label'] = le.fit_transform(df['Gender'])

print(df)
```

```python
# Simple Example: One Hot Encoding using sklearn

import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Sample dataset
data = {'City': ['Pune', 'Mumbai', 'Delhi', 'Pune']}
df = pd.DataFrame(data)
print("Original Data:")
print(df)

# Initialize the encoder
encoder = OneHotEncoder(sparse_output=False) #scikit-learn to return a dense NumPy array instead of a sparse matrix.

# Transform the data
encoded = encoder.fit_transform(df[['City']])

# Convert encoded data to DataFrame
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(['City']))

# Combine with original data
final_df = pd.concat([df, encoded_df], axis=1)

print("\nOne Hot Encoded Data:")
print(final_df)
```

```python
import pandas as pd

# Sample Data
data = {'City': ['Pune', 'Mumbai', 'Delhi', 'Pune']}
df = pd.DataFrame(data)

# One Hot Encoding in Binary Format
encoded_df = pd.get_dummies(df, columns=['City']).astype(int)

print(encoded_df)
```
