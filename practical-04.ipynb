#Use NumPy/pandas to calculate mean, median, mode, variance, std dev, correlation, and covariance.

## Mean

```python
import numpy as np
import pandas as pd

data = [10, 20, 30, 40, 50]
mean_np = np.mean(data)
mean_pd = pd.Series(data).mean()
print('NumPy Mean:', mean_np)
print('Pandas Mean:', mean_pd)
```

## Median

```python
import pandas as pd
import numpy as np
data = [1, 2, 2, 3, 4]
median_np = np.median(data)
median_pd = pd.Series(data).median()
print('NumPy Median:', median_np)
print('Pandas Median:', median_pd)
```

## Mode

```python
from scipy import stats
data = [1, 2, 2,5, 3, 4]
mode_val = stats.mode(data)
print('SciPy Mode:', mode_val)

```

```python
import pandas as pd

data = [1, 2, 2, 3, 4]
mode_val = pd.Series(data).mode()
print(mode_val)
```

## varience

```python
import pandas as pd
import numpy as np
data = [1, 2, 2, 3, 4]
var_np = np.var(data)
var_pd = pd.Series(data).var()
print('NumPy Variance:', var_np)
print('Pandas Variance:', var_pd)
```

## Standard Deviation


```python
import numpy as np
import pandas as pd
data1 = [10, 20, 30, 40, 50]
std_np = np.std(data)
std_pd = pd.Series(data).std()
print('NumPy Std Dev:', std_np)
print('Pandas Std Dev:', std_pd)
```

## Correlation

```python
import pandas as pd
df = pd.DataFrame({'data1': [10, 20, 30, 40, 50],
                   'data2': [15, 25, 35, 45, 55]})
correlation = df['data1'].corr(df['data2'])
print('Correlation coefficient:', correlation)

```

## Covariance


```python
import numpy as np

data1 = [10, 20, 30, 40, 50]
data2 = [15, 25, 35, 45, 55]

cov_matrix = np.cov(data1, data2)
covariance = cov_matrix[0, 1]  # covariance between data1 and data2
print('Covariance:', covariance)
```
