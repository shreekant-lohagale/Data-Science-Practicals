# Types of EDA Analysis
1. Univariate Analysis

```python
plt.Figure(figsize=(6,4))
sns.countplot(x="Survived", data=df)
plt.title("Survived Count (0 = No, 1 = Yes)")
plt.show()
```

```python
plt.Figure(figsize=(6,4))
sns.countplot(x="Sex", data=df)
plt.title("Sex Count (0 = No, 1 = Yes)")
plt.show()
```

```python
plt.Figure(figsize=(6,4))
sns.histplot(df["Age"].dropna(),bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.Figure(figsize=(8,5))
sns.histplot(df["Fare"].dropna(),bins=30, kde=True, color="purple")
plt.title("Fare Distribution")
plt.show()
```

```python
plt.Figure(figsize=(6,4))
sns.countplot(x="Pclass", data=df, palette="Set2")
plt.title("Passenger Class Distribution")
plt.show()
```

2. Bivariate Analysis

```python
plt.Figure(figsize=(6,4))
sns.countplot(x="Sex", hue="Survived", data=df, palette="Set1")
plt.title("Sex Count (0 = No, 1 = Yes)")
plt.show()
```

```python
plt.Figure(figsize=(6,4))
sns.countplot(x="Pclass", hue="Survived", data=df, palette="Set1")
plt.title("Sex Count (0 = No, 1 = Yes)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Survived", y="Age", data=df, palette="coolwarm")
plt.title("Age VS Survived")
plt.show()
```
